```python


num_colors = len(colors)
cols = 6
rows = math.ceil(num_colors / cols)

plt.figure(figsize=(18, rows * 1.2))

for idx, (name, hex_value) in enumerate(colors.items()):
    plt.subplot(rows, cols, idx + 1)
    plt.imshow([[mcolors.to_rgb(hex_value)]])
    plt.xticks([])
    plt.yticks([])
    plt.title(f"{name}\n{hex_value}", fontsize=8)

plt.tight_layout()
plt.show()
```
